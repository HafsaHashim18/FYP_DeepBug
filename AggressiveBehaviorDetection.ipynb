# Importing Libraries

```python
import PIL.Image
import torch
import numpy as np
import argparse
import joblib
import cv2
import torch.nn as nn
import torch.nn.functional as F
import time
import cnn_models
import albumentations
from torchvision.transforms import transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import utils

#using GPU Cuda
device = 'cuda:0'
```

# Breaking video into frames 

```python
# we will use OpenCV to read the video frames.
# And while training we have read the images using PIL.
# we will test our model by giving PIL format images to the model.
# The main reason for this is because of the difference between the RGB (Red, Green, Blue) and BGR (Blue, Green, Red) color formats of PIL and OpenCV.
# We have four command line arguments while executing the test.py file.

# construct the argument parser
ap = argparse.ArgumentParser()
ap.add_argument('-m', '--model', required=True,
                help="path to trained serialized model")
ap.add_argument('-l', '--label-bin', required=True,
                help="path to  label binarizer")
ap.add_argument('-i', '--input', required=True,
                help='path to our input video')
ap.add_argument('-o', '--output', required=True, type=str,
                help='path to our output video')
args = vars(ap.parse_args())


# Load the Binarized Labels, Prepare the Model, and Define the Image Augmentations
print('Loading model and label binarizer...')
lb = joblib.load(args['label_bin'])
model = cnn_models.CustomCNN().cuda()
model.to(device)
print('Model Loaded...')

# load the trained model and label binarizer from disk
model.load_state_dict(torch.load(args['model']))
print('Loaded model state_dict...')
aug = albumentations.Compose([
    albumentations.Resize(224, 224),
])
print("PRINTING MODEL", model)

# layers = list(model._modules['features']) + utils.toconv(list(model._modules['classifier']))
# L = len(layers)
#
# print("L=", L)
# For the augmentation, we will only be resizing the images into 224224 dimensions.

# Capturing the Video using OpenCV
cap = cv2.VideoCapture(args['input'])
if (cap.isOpened() == False):
    print('Error while trying to read video. Plese check again...')
    
# get the frame width and height
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# define codec and create VideoWriter object
out = cv2.VideoWriter(str(args['output']), cv2.VideoWriter_fourcc(*'mp4v'), 30, (frame_width, frame_height))

# We will read the video frame-by-frame until there are no more frames present.
# Then we will treat each frame as an image and carry out the predictions on each frame.
# read until end of video

output_array = []
pilimg_array = []
max = 0
max_ind = 0
while (cap.isOpened()):
    
    # capture each frame of the video
    ret, frame = cap.read()
    if ret == True:
        model.eval()
        with torch.no_grad():
            
            # convert to PIL RGB format before predictions
            pil_image = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            pil_image = aug(image=np.array(pil_image))['image']
            pil_image = np.transpose(pil_image, (2, 0, 1)).astype(np.float32)
            pil_image = torch.tensor(pil_image, dtype=torch.float).cuda()
            pil_image = pil_image.unsqueeze(0)
            pilimg_array.append(pil_image)

            outputs = model(pil_image)
            _, preds = torch.max(outputs.data, 1)

        cv2.putText(frame, lb.classes_[preds], (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 200, 0), 2)
        
        output_array.append((outputs.data))
        cv2.imshow('image', frame)
        out.write(frame)
        
        # press `q` to exit
        if cv2.waitKey(27) & 0xFF == ord('q'):
            break
    else:
        break
        
# release VideoCapture()
cap.release()

# close all frames and video windows
cv2.destroyAllWindows()
```

# Extracting Maximum frame

```python
# Extracting the maximum aggressive frame
maximum = 0
max_ind = 0
for i, frame in enumerate(output_array):
    if frame[0][0] > maximum:
        maximum = frame[0][0]
        max_ind = i


# Getting the maximum frame from array and converting to RGB for the purpose of displaying
pil_image = pilimg_array[max_ind].cuda()
img = pilimg_array[max_ind].squeeze(0)
im = (transforms.ToPILImage()(img).convert("RGB"))

print("\nInput frame shape\n", pil_image.shape)
X = pil_image

```

# Model Layers

```python
model.eval()
```

```python
# Calculating model layers
layers = [module for module in model.modules() if not isinstance(module, torch.nn.Sequential)][1:]
print("Total Layers in model ",len(layers))
L = len(layers)

```

# Forward Activations

```python
# Image as 1st activation and 7 additional Empty activation arrays
A = [X] + [None] * L
```

```python
# First four conv2d layers
for l in range(L-3): 
    A[l+1] = layers[l].forward(A[l])
    print("Activation of layer",l,":",A[l].shape)
```

```python
l = 4
# Layer 5 is the first fully connected layer so Reshaping is required
bs, _, _, _ = A[l].shape
A[l] = F.adaptive_avg_pool2d(A[l], 1).reshape(bs, -1)
A[l+1] = layers[l].forward(A[l])

print("Activation of layer",l,":",A[l].shape)
```

```python
l += 1
A[l+1] = layers[l].forward(A[l])
print("Activation of layer",l,":",A[l].shape)

l += 1
A6 = (A[l])

#NON AGRESSIVE INDEX MADE ZERO at the last activation layer
A6[0][1] = 0
A[l+1] = A6

print("Activation of layer",l,":",A[l].shape)
```

# Relevances

```python
R = [None]*L + [A[7].data]
```

```python
# Relevance layer 6

l=6
A[l] = (A[l].data).requires_grad_(True)
R[l] = R[l + 1]
print ("Relevance of layer 6 : ",R[l],R[l].shape)
```

```python
l =5

A[l] = (A[l].data).requires_grad_(True)

if l <= 2:       rho = lambda p: p + 0.25 * p.clamp(min=0); incr = lambda z: z + 1e-9
if 3 <= l <= 4: rho = lambda p: p;                       incr = lambda z: z + 1e-9 + 0.25 * (
        (z ** 2).mean() ** .5).data
if l >= 5:       rho = lambda p: p;                       incr = lambda z: z + 1e-9

nl = utils.newlayer(layers[l], rho)
f = nl.forward(A[l])
z = incr(f)  # step 1
s = (R[l + 1]/ z).data  # step 2
(z * s).sum().backward();
c = A[l].grad  # step 3
R[l] = (A[l] * c).data  # step 4

print ("Relevance of layer 5 : ",R[l].shape)
```

```python
l =4

A[l] = (A[l].data).requires_grad_(True)

if l <= 2:       rho = lambda p: p + 0.25 * p.clamp(min=0); incr = lambda z: z + 1e-9
if 3 <= l <= 4: rho = lambda p: p;                       incr = lambda z: z + 1e-9 + 0.25 * (
        (z ** 2).mean() ** .5).data
if l >= 5:       rho = lambda p: p;                       incr = lambda z: z + 1e-9

nl = utils.newlayer(layers[l], rho)
f = nl.forward(A[l])
z = incr(f)  # step 1
s = (R[l + 1]/ z).data  # step 2
(z * s).sum().backward();
c = A[l].grad  # step 3
R[l] = (A[l] * c).data  # step 4

print ("Relevance of layer 4 : ",R[l].shape)
```

```python
l =3

A[l] = (A[l].data).requires_grad_(True)

if l <= 2:       rho = lambda p: p + 0.25 * p.clamp(min=0); incr = lambda z: z + 1e-9
if 3 <= l <= 4: rho = lambda p: p;                       incr = lambda z: z + 1e-9 + 0.25 * (
        (z ** 2).mean() ** .5).data
if l >= 5:       rho = lambda p: p;                       incr = lambda z: z + 1e-9

z = incr(utils.newlayer(layers[l], rho).forward(A[l]))  # step 1

R_prev=R[l+1]

# Reshaping of dimensions needs to be done in backward pass between fully connected and convolutional layer

R_new = np.arange(1*128*210*210)
R_new = R_new.reshape(1,128,210,210)

# Redistributing mean value that we go from the Avgpooling in forward pass

for i in range (128):
    R_new[0][i]=R_prev[0][i].item()
    for j in range (210):
        for k in range(210):
            R_new[0][i][j][k]=R_prev[0][i].item()
            
R[l+1]=torch.FloatTensor(R_new)

s = (R[l + 1].cuda() / z).data  # step 2
(z * s).sum().backward();
c = A[l].grad  # step 3
R[l] = (A[l] * c).data  # step 4

print ("Relevance of layer 3 : ",R[l].shape)
```

```python
l =2

A[l] = (A[l].data).requires_grad_(True)

if l <= 2:       rho = lambda p: p + 0.25 * p.clamp(min=0); incr = lambda z: z + 1e-9
if 3 <= l <= 4: rho = lambda p: p;                       incr = lambda z: z + 1e-9 + 0.25 * (
        (z ** 2).mean() ** .5).data
if l >= 5:       rho = lambda p: p;                       incr = lambda z: z + 1e-9

z = incr(utils.newlayer(layers[l], rho).forward(A[l]))  # step 1
s = (R[l + 1] / z).data  # step 2
(z * s).sum().backward();
c = A[l].grad  # step 3
R[l] = (A[l] * c).data  # step 4

print ("Relevance of layer 2 : ",R[l].shape)
```

```python
l =1

A[l] = (A[l].data).requires_grad_(True)

if l <= 2:       rho = lambda p: p + 0.25 * p.clamp(min=0); incr = lambda z: z + 1e-9
if 3 <= l <= 4: rho = lambda p: p;                       incr = lambda z: z + 1e-9 + 0.25 * (
        (z ** 2).mean() ** .5).data
if l >= 5:       rho = lambda p: p;                       incr = lambda z: z + 1e-9

z = incr(utils.newlayer(layers[l], rho).forward(A[l]))  # step 1
s = (R[l + 1] / z).data  # step 2
(z * s).sum().backward();
c = A[l].grad  # step 3
R[l] = (A[l] * c).data  # step 4

print ("Relevance of layer 1 : ",R[l].shape)
```

```python
# Better Highlighting the most significant portions of the image at 0th layer while recreating heatmap

A[0] = (A[0].data).requires_grad_(True)
mean = torch.Tensor([0.485, 0.456, 0.406]).reshape(1, -1, 1, 1)
std = torch.Tensor([0.229, 0.224, 0.225]).reshape(1, -1, 1, 1)
x=(0-mean)/std
y=A[0].data*0
z=(1-mean)/std
lb = (y.cuda()+x.cuda()).requires_grad_(True)
hb = (y.cuda()+z.cuda()).requires_grad_(True)

z = layers[0].forward(A[0]) + 1e-9                                     # step 1 (a)
z -= utils.newlayer(layers[0],lambda p: p.clamp(min=0)).forward(lb)    # step 1 (b)
z -= utils.newlayer(layers[0],lambda p: p.clamp(max=0)).forward(hb)    # step 1 (c)
s = (R[1]/z).data                                                      # step 2
(z*s).sum().backward(); c,cp,cm = A[0].grad,lb.grad,hb.grad            # step 3
R[0] = (A[0]*c+lb*cp+hb*cm).data  

print ("Relevance of layer 0 : ",R[0].shape)
```
